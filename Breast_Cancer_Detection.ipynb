## Correlation matrix and heatmap

```python
dataset2 = dataset.drop(columns='diagnosis_M')
dataset2.head()
# diagnosis_M is a independent variable and except it all are the dependent variables.
# so we are findind a correlation between dependent and independent variables.
```

```python
dataset2.corrwith( dataset['diagnosis_M']).plot.bar(
    figsize=(20,10),title='Correlated with diagnosis_M',rot=45,grid=True
)
```

```python
#Correlation Matrix
corr=dataset.corr()
```
